```python
df.columns=['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate', 'Area']
```

```python
plt.figure(figsize=(12,10))

sns.heatmap(df.corr().round(2), cmap='Spectral', annot=True)

plt.show()
```

```python
df["Area"].value_counts()
```

```python
# Visualising the whole dataset

plt.rcParams['figure.figsize']=20,20

sns.pairplot(df, hue="Area", palette="plasma")

plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df["month_int"] =  df["Date"].dt.month
df['month'] =  df['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
print("\nState with the Highest Unemployment Rate(%): ", df["States"].value_counts().idxmax())
print("\nState with the Lowest Unemployment Rate(%): ", df["States"].value_counts().idxmin())
print("\n"+"-----------"*10)
print("\nMonth with the Highest Unemployment Rate(%): ", df["month"].value_counts().idxmax())
print("\nMonth with the Lowest Unemployment Rate(%): ", df["month"].value_counts().idxmin())
```

```python
#top 10 states with highest unemployment

df1 = df[["States", "Estimated Unemployment Rate"]].groupby("States").sum().sort_values(by="Estimated Unemployment Rate", ascending=False)
```

```python
df1.head(10)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

df1[:10].plot(kind="bar", color="blue",figsize=(20,10),ax=ax)

ax.set_title("Top 10 States with Highest Unemployment")
ax.set_xlabel("States")
ax.set_ylabel("Number of people unemployed in %")
plt.xticks(rotation=0)

plt.show()
```

```python
df2 = df[["month","Estimated Unemployment Rate"]].groupby("month").sum().sort_values(by="Estimated Unemployment Rate", ascending  =False)
```

```python
df2.head(10)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

df2[:12].plot(kind="bar", color="yellow",figsize=(20,10),ax=ax)

ax.set_title("Months with highest unemployment")
ax.set_xlabel("Region")
ax.set_ylabel("Number of People Unemployed (in %)")
plt.xticks(rotation=0)

plt.show()
```
